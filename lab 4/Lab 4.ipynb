We have a knapsack of capacity weight C (a positive integer) and n types of objects.
Each object of the ith type has weight wi and profit pi (all wi and all pi are positive
integers, i = 0, 1, â€¦, n-1). There are unlimited supplies of each type of objects. Find
the largest total profit of any set of the objects that fits in the knapsack.
Let P(C) be the maximum profit that can be made by packing objects into the knapsack
of capacity C.
### Part (1) Give a recursive definition of the function P(C).

```python
## let profit be array of item value
## let weight be array of item weight
## let capacity be max capacity
## let itemCount be number of items

def unboundedKnapsack(capacity, itemCount, profit, weight, debug = False):
 
    # memo[i] stores the maximum profit attainable with knapsack capacity i.
    memo = [0 for i in range(capacity + 1)]
 
    for c in range(capacity + 1):
        for item in range(itemCount):
            if (weight[item] <= c):
                memo[c] = max(memo[c], memo[c - weight[item]] + profit[item])
        debug and print(c, memo)
 
    return memo[capacity]
```


#### Therefore ,Recursive Definition, C = capacity
### Since P(C) is a function to return the max profit made by checking all objects against each P(C),
**Base Case:** P(C) = 0, (C=0)

**Recursive Definition:** P(C) = max(profit[0]+P(C-weight[0]),profit[1]-P(C+weight[1],.....,profit[n-1]+P(C-weight[n-1])), where weight[i] <= C

### Part (3) Bottom up algorithm for unbounded knapsack problem

```python
def unboundedKnapsack(capacity: int, itemCount: int, profit: list, weight:list, debug = False):
 
    # memo[i] stores the maximum profit attainable with knapsack capacity i.
    memo = [0 for i in range(capacity + 1)]
 
    for c in range(capacity + 1):
        for item in range(itemCount):
            if (weight[item] <= c):
                memo[c] = max(memo[c], memo[c - weight[item]] + profit[item])
        debug and print(c, memo)
 
    return memo[capacity]
```

```python
capacity = 14
weight_part2 = [4,6,8]
profit_part2 = [7,6,9]
 
unboundedKnapsack(capacity, len(profit_part2), profit_part2, weight_part2, True)
```
